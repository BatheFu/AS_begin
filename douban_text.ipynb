```python
import requests
import json

dat = pd.read_csv('new.csv', encoding="utf-8")
# Connect to the mongoDB

from pymongo import MongoClient
con = MongoClient()
db = con['scrapydata']
posts = db['topic']
con.list_database_names()



for i in range(928):
    s = requests.sessions.Session()
    topic_url = dat.iloc[i, :][4]
    article_num = dat.iloc[i, :][1]
    jsonurl = dat.iloc[i, :][-1]
    jsonurl = jsonurl + '/items'
    print(topic_url, '\n', article_num, '\n', jsonurl)
    # Define Headers and refined params
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/75.0.3770.142 Safari/537.36',
        'Referer': str(topic_url),

    }
    if article_num <= 50:
        sendtime = 20
    else:
        sendtime = 50
    
    data = {'sort': 'hot',
            'start': 0,
            'count': sendtime,
            'status_full_text': 1,
            }
    

    # Send the requests
    r = s.get(jsonurl, headers=headers, params=data)
    if r.status_code == 200:
        print("Success on Url", topic_url)
    else:
        print("error on processing url{}".format(topic_url))
    # Get the json file
    j = r.json()
    # Get the text
    text = []
    author_id = []
    author_name = []
    topic_name = []
    update_time = []
    for index in range(int(sendtime)):
        if 'status' in j['items'][index]['target']:
            text.append(j['items'][index]['target']['status']['text'])
            author_id.append(j['items'][index]['target']['status']['author']['id'])
            author_name.append(j['items'][index]['target']['status']['author']['name'])
            topic_name.append(j['items'][index]['target']['status']['topic']['name'])
            update_time.append(j['items'][index]['target']['status']['create_time'])
        else:
            text.append(j['items'][index]['target']['abstract'])
            author_id.append(j['items'][index]['target']['author']['uid'])
            author_name.append(j['items'][index]['target']['author']['name'])
            topic_name.append(j['items'][index]['target']['title'])
            update_time.append(j['items'][index]['target']['update_time'])

    text_list = list()
    keys = ['author_id','author_name', 'text', 'topic_name','update_time']
    for index in range(len(author_id)):
        values = [author_id[index], author_name[index], text[index], topic_name[index], update_time[index]]
        text_dic = dict(zip(keys, values))
        text_list.append(text_dic)
    posts.insert_many(text_list)
```
